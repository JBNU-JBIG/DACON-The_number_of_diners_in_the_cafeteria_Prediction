### 1.1 날짜와 요일 
날짜를 `datetime` 형식으로 바꾸면 여러모로 편하기 때문에 바꿔줍니다. 그리고 `dow` (day of week; 요일) 정보를 숫자로 만들어줍니다. 

```python
def to_datetime(df, date):
    df['date'] = pd.to_datetime(df[date])
    df['dow'] = pd.to_datetime(df[date]).dt.weekday + 1
to_datetime(train, 'date'); to_datetime(test, 'date')
```

### 1.2 메뉴명
이 데이터셋의 핵심은 메뉴명 정보라고 할 수 있습니다. 이를 어떻게 전처리하는지부터 시작해서 어떤 모델을 사용할지까지, 점심과 저녁 메뉴에 따라 결정할 수 있을 것 같습니다. 

일차적으로 점심 세부 메뉴명을 구분하여 각 칼럼에 저장하는 방법으로 전처리를 해보았는데, 이렇게 하는게 맞는지, 이를 어떻게 모델에서 활용할지는 좀 더 고민해봐야 할 것 같습니다. 다른 분들의 의견 환영합니다! 

```python
# 일별 점심메뉴를 작은 리스트로 갖고 있는 큰 리스트 (lunch) 만들기 
lunch = []
for day in range(len(train)):
    tmp = train.iloc[day, 8].split(' ') # 공백으로 문자열 구분 
    tmp = ' '.join(tmp).split()    # 빈 원소 삭제

    search = '('   # 원산지 정보는 삭제
    for menu in tmp:
        if search in menu:
            tmp.remove(menu)
    
    lunch.append(tmp) 
```

```python
# 확인 
lunch[0:5]
```

```python
lunch[1065:1070]
```

* 이런... 점심메뉴를 살펴보니, 처음에는 밥/국/반찬3개/사이드(요구르트 등)/김치 순서로 기록되어 있다가, 어느 날 갑자기 사이드(샐러드 등)와 김치의 순서가 뒤바뀝니다. 이를 고려하여 메뉴명을 추출해야겠네요.
   + `1067번` 행부터 바뀌는 것을 확인할 수 있습니다. 추가로 어떤 날짜에서 순서가 바뀌는지 확인을 해보겠습니다.   

```python
np.array(train[ (train.index > 1064) & (train.index < 1069)][['date', 'ln']])
```

`2020-07-01`부터 순서가 바뀌는군요. `2020-06-13`부터 `2020-06-30`까지는 구내식당이 영업을 하지 않았나 봅니다 (데이터가 아예 없으니까요). 그 이유가 뭐든, 그 이후부터 메뉴명 순서가 바뀝니다. 

메뉴는 밥, 국, 반찬 1-3, 김치, 사이드로 구분해보겠습니다. 
  * 여기서 사이드는 다른 메뉴들을 제외한 메뉴이며, 주로 `요구르트`나 `샐러드`류가 있습니다. 

```python
# lunch train data에 메뉴명별 칼럼 만들기 (밥, 국, 반찬1-3)
bob = []; gook = []; banchan1 = []; banchan2 = []; banchan3 = []; kimchi = []; side = []
for i, day_menu in enumerate(lunch):
    bob_tmp = day_menu[0]; bob.append(bob_tmp)
    gook_tmp = day_menu[1]; gook.append(gook_tmp)
    banchan1_tmp = day_menu[2]; banchan1.append(banchan1_tmp)
    banchan2_tmp = day_menu[3]; banchan2.append(banchan2_tmp)
    banchan3_tmp = day_menu[4]; banchan3.append(banchan3_tmp)

    if i < 1067:
        kimchi_tmp = day_menu[-1]; kimchi.append(kimchi_tmp)
        side_tmp = day_menu[-2]; side.append(side_tmp)
    else: 
        kimchi_tmp = day_menu[-2]; kimchi.append(kimchi_tmp)
        side_tmp  = day_menu[-1]; side.append(side_tmp)
```

```python
train_ln = train[['date', 'dow', 'employees', 'dayoff', 'bustrip', 'ovtime', 'remote', 'ln', 'target_ln']]
train_ln['bob'] = bob
train_ln['gook'] = gook
train_ln['banchan1'] = banchan1; train_ln['banchan2'] = banchan2; train_ln['banchan3'] = banchan3
train_ln['kimchi'] = kimchi 
train_ln['side'] = side
```

### 2.3 변수 간 상관관계 

```python
train.head(2)
```

```python
df = train[['target_ln', 'target_dn', 'employees', 'dayoff', 'bustrip', 'ovtime', 'remote']]
mask = np.triu(np.ones_like(df.corr(), dtype=np.bool))
plt.rcParams['font.size'] = 15

fig, ax = plt.subplots(figsize=(16, 5))
sns.heatmap(df.corr(), 
            annot=True, 
            cmap="BrBG", 
            mask = mask)
ax.set_title('Correlation Heatmap', pad = 10)
plt.show()
```

저희가 예측하고 싶은 변수(종속변수)는 점심과 저녁 이용자수입니다. 다시 말해, `target_ln` (점심)와 `target_dn` (저녁) 이죠. 
* 일반적으로, 상관계수의 크기가 0.4 이상이면 상관성이 있다고 볼 수 있습니다. 
* 먼저, 점심을 보죠. 
  + `target_ln`에 해당하는 열을 보면, 이 변수와 가장 상관관계가 높은 독립변수는 **`ovtime (0.54)` (시간외근무자수)**이고, 그 다음으로는 **`bustrip (-0.51)` (출장자수)** 입니다. 
     + 물론, 시간외근무자수는 양의 관계이고 출장자수는 음의 관계입니다. 각각 회사 내에 직원이 많이 있느냐와 적게 있느냐에 영향을 끼치는 요인이니까요. 
     + 특이한 것은, 시간외근무자라는 것은 야근을 한다는 것인데, 저녁뿐만이 아니라 점심 이용자수와도 상관관계를 보인다는 점입니다. 
  + `dayoff (-0.39)`(휴가자수) 또한 출장자수와 비슷하게, 그 값이 크면 회사에 출근한 직원수가 적다는 의미이므로 음의 관계를 갖고 있습니다. `dayoff`의 상관계수의 크기는 `bustrip`의 계수보다는 작지만요. 
* 다음은 저녁입니다.  
  + 사실 점심 때와 크게 다르지는 않습니다. 차이가 있다면 애초에 회사 내에 없었던 직원들의 수 (`bustrip`, `dayoff`)와의 상관관계가 점심 때보다는 약해졌다는 것입니다. 
  + 이보다 더 중요하다고 생각되는 점은 **`ovtime`**의 상관계수가 `0.57`로 조금 커진 것입니다. 
      + 이는 충분히 예상할 수 있었던 결과입니다. 야근하는 직원수가 많으면 당연히 저녁 이용자수가 많아질 것이니까요. 

이번에는 각 변수별로 산점도를 그려보겠습니다. 

```python
fig, ax = plt.subplots(figsize = (18, 8), ncols = 4, nrows = 2, sharey=True)
plt.rcParams['font.size'] = 12
sns.color_palette("Paired")
train_features = ['employees', 'dayoff', 'bustrip', 'ovtime', 'employees', 'dayoff', 'bustrip', 'ovtime']
for i, feature in enumerate(train_features):
    row = int(i/4)
    col = i%4 
    if i < 4:
        sns.regplot(x=feature, y = 'target_ln', data = train, ax = ax[row][col], color = 'salmon', marker = '+')
    else: 
        sns.regplot(x=feature, y = 'target_dn', data = train, ax = ax[row][col], color = 'skyblue', marker = '+')
```

상관관계를 히트맵으로 그리든, 산점도로 그리든 결과는 같으니 산점도에 대한 설명은 생략하도록 하겠습니다. 

### 2.4 월별 & 요일별 패턴
위에서 변수 간 상관관계를 보았는데요, 이번에는 월별 & 요일별 점심과 저녁 이용자수의 패턴에 집중해보죠. 이 데이터셋에는 날짜 정보가 있으므로, 이를 이용해보겠습니다. 

#### (1) 점심 및 저녁 이용자수 
* 월별로 점심과 저녁 이용자수가 다를 것 같습니다. 일이 많고 바쁜 달에는 밖에서 먹기보다는 회사 구내식당에서 끼니를 해결할 것 같기 때문이죠. 야근도 더 많이 할 것 같고요. 
* 요일에 따라서도 점심과 저녁 이용자수가 달라질 것 같습니다. 예를 들어, 저라면 금요일 저녁은 회사에서 먹기보다는 가족과 함께 집에서 먹고 싶기 때문이죠. 반대로 월요일은 보통 일이 많은 날이니, 구내식당에서 자주 먹을 것 같습니다. 

```python
train.head(3)
```

히트맵을 이용하여 월별 & 요일별 점심과 저녁 이용자수의 평균을 시각화해보겠습니다. 

```python
tmp = train[['date', 'dow', 'employees', 'dayoff', 'bustrip', 'ovtime', 'remote', 'target_ln', 'target_dn']]
tmp['month'] = tmp['date'].dt.strftime("%m")

tmp_ln = tmp.groupby(['dow', 'month'])['target_ln'].mean().reset_index().pivot('dow', 'month', 'target_ln')
tmp_dn = tmp.groupby(['dow', 'month'])['target_dn'].mean().reset_index().pivot('dow', 'month', 'target_dn')
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (15, 4))

sns.heatmap(tmp_ln, cmap='RdYlGn_r', ax=ax[0])
ax[0].set_title('Lunch', pad = 12)
sns.heatmap(tmp_dn, cmap='RdYlGn_r', ax=ax[1])
ax[1].set_title('Dinner', pad = 12)

plt.show()
```

점심 
* **월요일**에 압도적으로 이용자수가 많습니다. 유일하게 빨갛네요. 늘 1000명보다 많다는 뜻이겠죠. 
  + 다른 요일에 비해 월요일에 점심 이용자수가 많은 것은 1년 내내 보여지는 패턴입니다. 
  + 그렇지만 같은 월요일이라도 **2월과 3월**에는 이용자수가 더 많네요. 
* 화요일과 수요일에도 대략 1000명 정도의 이용자수가 있는 달이 있는데, 역시나 2월과 3월입니다. 
  + LH는 2-3월에 가장 일이 몰려있나 봅니다. 물론, 일이 많으면 구내식당에서 점심을 먹는다는 가설이 성립해야만 가능한 설명이죠. 
* 목요일과 금요일은 점심 이용자수가 적습니다. 특히 금요일은 거의 짙은 초록색 뿐이네요. 직원들은 금요일에는 저녁뿐만이 아니라 점심도 회사 내에서 먹기 싫나 봅니다 (또 다른 이유는 출장자수인데, 이는 이후에 확인해보겠습니다). 

저녁 
* 저녁 이용자수의 전체 인원은 점심 이용자수보다 적다는 것을 먼저 알아야 합니다. 각 히트맵의 좌측 scale을 보면, 점심은 `700-1200(명)`이지만 저녁은 `350-600(명)`이라는 것을 알 수 있죠? 다만, 저는 여기서 요일과 월별 패턴을 보고 싶은 것이니, 이용자수 자체는 자세히 언급하지 않겠습니다. 
* 점심과 마찬가지로, 저녁 이용자수는 **월요일**이 가장 많아보입니다. 
* 점심과는 다르게, **화요일**에도 월요일과 비슷한 수준으로 이용자수가 많네요. 
   + 월, 화요일에는 야근하는 직원들이 많나 봅니다 (이후에 확인해볼 것입니다). 
* 그리고 예상외로 수요일에 저녁 이용자수가 매우 적은데.... 아, 이것은 **'자기 개발의 날'** 때문이겠군요. 저녁 메뉴가 없는 데이터는 삭제한 뒤, 다시 저녁 히트맵을 시각화해보겠습니다. 

```python
# 자기개발의 날 삭제 
idx = train[train.target_dn == 0].index
tmp = train.drop(idx)
tmp['month'] = tmp['date'].dt.strftime("%m")
tmp_dn2 = tmp.groupby(['dow', 'month'])['target_dn'].mean().reset_index().pivot('dow', 'month', 'target_dn')

fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (15, 4))

sns.heatmap(tmp_ln, cmap='RdYlGn_r', ax=ax[0])
ax[0].set_title('Lunch', pad = 12)
sns.heatmap(tmp_dn2, cmap='RdYlGn_r', ax=ax[1])
ax[1].set_title('Dinner - NEW', pad = 12)

plt.show()
```

점심은 동일한 히트맵입니다.  
  
저녁 
* '자기 개발의 날`을 제외했는데도, 수요일에는 여전히 저녁 이용자수가 적은 편입니다.  
  + 생각해보니 월, 화요일에 저녁을 구내식당에서 먹어야 할만큼 야근을 많이 했다면, 수요일만큼은 회사에서 저녁을 먹기 싫을 것 같네요. 야근 자체를 안하던지, 밖에서 저녁을 먹고 싶을 것 같습니다. 
* 금요일에도 예상대로 적습니다. 
* **2-3월**에는 확실히 야근하는 직원들의 수가 많나 봅니다. 각 요일에서 월별로 비교해보면, 확실히 2-3월의 저녁 이용자수가 늘 많은 편입니다. 

점심과 저녁 이용자수 모두 **대체로 요일별 패턴이 가장 극명하게 보입니다.** 월별 차이보다는요. 이는 곧 구내식당 이용자수를 예측할 때 요일이 중요한 요인이라고 할 수 있습니다. 

#### (2) 회사에 있는 직원 수 
이번에는 요일 & 월별로 회사 내에 있는 직원수가 얼마나 다른지 확인해보겠습니다. 이들은 잠재적으로 점심과 저녁을 먹을 사람들이지요.  
 * 회사 내에 있는 직원수는 `employees - (dayoff + bustrip + remote)` 으로 구할 수 있을 것입니다. 다시 말해, 본사정원수 - (본사휴가자수 + 본사출장자수 + 재택근무자수) 입니다. 
 * 이를 구하기 전에, 휴가자수(`dayoff`), 출장자수(`bustrip`), 시간외근무자수(`ovtime`), 재택근무자수(`remote`)를 요일별 & 월별로 시각화해보겠습니다. 
   +  다만, 재택근무수를 확인해보니 (아래), 코로나 이전인 2019년 까지는 0명이었습니다. 그러므로, 재택근무자수는 2020년의 데이터만 시각화하겠습니다. 
   

```python
before = train['date'].dt.year < 2020
after = train['date'].dt.year >= 2020

train[before]['remote'].value_counts()  # 코로나19 전 재택근무자수 
```

```python
before = train['date'].dt.year < 2020
after = train['date'].dt.year >= 2020

def heatmap_viz(df): 
    df['month'] = df['date'].dt.strftime("%m")
    before = df['date'].dt.year < 2020
    after = df['date'].dt.year >= 2020

    tmp_dayoff = df.groupby(['dow', 'month'])['dayoff'].mean().reset_index().pivot('dow', 'month', 'dayoff')
    tmp_bustrip = df.groupby(['dow', 'month'])['bustrip'].mean().reset_index().pivot('dow', 'month', 'bustrip')
    tmp_ovtime = df.groupby(['dow', 'month'])['ovtime'].mean().reset_index().pivot('dow', 'month', 'ovtime')
    tmp_remote_after = df[after].groupby(['dow', 'month'])['remote'].mean().reset_index().pivot('dow', 'month', 'remote')

    fig, ax = plt.subplots(nrows = 1, ncols = 4, figsize = (30, 5), sharey = True)

    sns.heatmap(tmp_dayoff, cmap='Oranges', ax=ax[0])   #1 
    ax[0].set_title('Dayoff', pad = 12)
    sns.heatmap(tmp_bustrip, cmap='Greens', ax=ax[1])   #2 
    ax[1].set_title('Business Trip', pad = 12)
    sns.heatmap(tmp_ovtime, cmap='Blues', ax=ax[2])   #3
    ax[2].set_title('Overtime', pad = 12)
    sns.heatmap(tmp_remote_after, cmap='Purples', ax=ax[3])   # 4
    ax[3].set_title('Remote (2020-21 only)', pad = 12)
   
    plt.show()

df = train[['date', 'dow', 'dayoff', 'bustrip', 'ovtime', 'remote']]
heatmap_viz(df)
```

휴가자수, 출장자수, 시간외근무자수, 재택근무자수는 확연히 다른 패턴을 보입니다. 상대적으로 그 인원수가 많은 요일 and/or 월은 다음과 같습니다. 
* 휴가자수 (`Dayoff`): 월/금, 7/8/12월 
* 출장자수 (`Business Trip`) : 금, 5/6/9/10/11월
   + 점심 이용자수가 금요일에 매우 적었던 것을 보면, 당연한 결과입니다. 출장을 한다는 것은 하루종일 회사 내에 없는 것이니까요. 
* 시간외근무자수 (`Overtime`) : 월/화/목, 2/3월 
   + 저녁 이용자수가 수, 금요일에 적은 편인 것을 생각해보면, 자연스러운 결과입니다. 
* 재택근무자수 (`Remote`) : 4/9/12월 

이제 회사 내에 있는 직원수를 계산한 뒤, 이를 위와 같은 방법으로 시각화해보겠습니다. 
  * 회사 내에 있는 직원수 = `employees - (dayoff + bustrip + remote)` 

```python
df = train[['date', 'dow', 'employees', 'dayoff', 'bustrip', 'ovtime', 'remote', 'target_ln', 'target_dn']]
df['in_office'] = df['employees'] - (df['dayoff'] + df['bustrip'] + df['remote'])
df['month'] = df['date'].dt.strftime("%m")
df.head(3)
```

```python
tmp = df.groupby(['dow', 'month'])['in_office'].mean().reset_index().pivot('dow', 'month', 'in_office')

# Heatmap
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 4))

sns.heatmap(tmp, cmap='YlGnBu', ax = ax[0])    # 1 
ax[0].set_title('No. of Employees in Office', pad = 12)

df_corr = df[['target_ln', 'target_dn', 'employees', 'dayoff', 'bustrip', 'ovtime', 'remote', 'in_office']]   # 2
mask = np.triu(np.ones_like(df_corr.corr(), dtype=np.bool))
sns.heatmap(df_corr.corr(), 
            annot=True, 
            cmap="BrBG", 
            mask = mask, 
            ax =ax[1])
ax[1].set_title('Correlation Heatmap ("in_office" added)', pad = 10)
plt.show()
```

회사 내 직원수 (왼쪽)
* 쉽게 예상할 수 있듯이, 금요일에는 상대적으로 회사 내에 있는 직원이 적습니다. 
   + 휴가자수와 출장자수가 금요일에 많았었죠. 
* 월별로 봐도 1-3월, 9-11월에는 본사 내에서 일하는 직원이 많아 보입니다.   

상관관계 (오른쪽) 
* `in_office`라는 변수를 새로 추가했기 때문에, 새롭게 상관관계를 시각화해보았습니다. 
* 회사 내에 있는 직원수를 의미하는 `in_office`는 점심과 저녁 이용자 간 상관계수가 순서대로 `0.29`와 `0.17`인데, 그렇게 강한 상관관계는 아닌 듯 합니다.
